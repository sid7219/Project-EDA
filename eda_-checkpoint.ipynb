```python
df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].median())   # Fills the NaN value in col. by filling median value in num col.
```

```python
df['Loan_Amount_Term']=df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].median())
```

```python
df['Credit_History']=df['Credit_History'].fillna(df['Credit_History'].median())
      
```

```python
df['Self_Employed']=df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])          # Fills the NaN value in col. by filling mode value in cat col.
```

```python
df['Gender']=df['Gender'].fillna(df['Gender'].mode()[0])
```

```python
df['Dependents']=df['Dependents'].fillna(df['Dependents'].mode()[0])
```

```python
'''
Plotting Graphs:

Univariate:
Nume Data → Histogram / KDE Plots and BoxPlot
Cat Dara → CountPlot

Bivariate:
Num vs Num → Scatter
Cat vs Num → Boxplot
Cat vs Cat → Countplot

Multivariate:
→ Heatmaps, Pairplot
'''
```

```python
df.describe(include='object')
```

```python
# Histogram and KDE Plot for LoanAmount, ApplicantIncome, CoapplicantIncome

plt.hist(df['LoanAmount'],bins=30)
plt.title("Histogram of LoanAmount")
plt.show()

sns.kdeplot(df['LoanAmount'])
plt.title("KDE Plot of LoanAmount")
plt.show()

plt.hist(df['ApplicantIncome'],bins=30)
plt.title("Histogram of ApplicantIncome")
plt.show()

sns.kdeplot(df['ApplicantIncome'])
plt.title("KDE Plot of ApplicantIncome")
plt.show()

plt.hist(df['CoapplicantIncome'],bins=30)
plt.title("Histogram of CoapplicantIncome")
plt.show()
```

```python
# BoxPlot 

sns.boxplot(x=df["LoanAmount"])
plt.show()

sns.boxplot(x=df["ApplicantIncome"])
plt.show()

sns.boxplot(x=df["CoapplicantIncome"])
plt.show()
```

```python
# CountPlot

sns.countplot(x=df['Gender'])

sns.countplot(x=df['Education'])

sns.countplot(x=df['Property_Area'])
```

```python
# ScatterPlot 

sns.scatterplot(x=df['ApplicantIncome'],y=df['LoanAmount'],hue=df['Gender'],style=df['Education'])
plt.show()
```

```python

sns.scatterplot(x=df['ApplicantIncome'],y=df['LoanAmount'],hue=df['Property_Area'])
plt.show()
```

```python

sns.boxplot(x=df['ApplicantIncome'],y=df['Gender'])
```

```python
# Heatmap- Shows how one nm feature moves with the other

corr=df.select_dtypes(include='number').corr()
sns.heatmap(corr,cmap='coolwarm',annot=True)
```
